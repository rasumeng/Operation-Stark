## What is this dataset good for?
The dataset is good for:
- Logistic Regression
    - Logistic Regression is used when the target is binary
    - Instead of predicting y directly it predicts: P(Y = 1|X)
    - Biggest Difference in Logistic regression is the Sigmoid function
        - The regression wraps the linear equation isnide a sigmoid function
        $\sigma(z) = \frac{1}{1 + e^{-z}}$
        - WHere: $y = b_0 + b_1x_1 + b_2x_2 + \dots + b_nx_n$
    - The sigmoid: keeps the output between 0 and 1 giving a true probability

- Random Forest
    - Instead of one decision tree I am building muany decision trees
    - Each tree:
        - Mkaes its own prediciton
        - looks at random subsets of data
        - looks at random subsets of features
    - Then:
        - Final prediction = majority vote (classification)
    - This is better because an indiviudal tree would overfit
    - A forest averages them → reducing overfitting.
- XGBoost (Extreme Gradient Boosting)
    - It builds trees sequentially unlike Random forest that builds them sequentially
        - Each new tree focuses on correcting mistakes of previous trees
- SVM
    - Tries to find the best boundary that separates two classes. 
    - Like plotting defaulters vs non-defaulters
    - Finds the line (or hyperplane) that:
        - Maximizes the margin between the two classes.
    - Margin = distance between boundary and closest points.
    - Powerful but less scalable than tree-based methods and is more math heavy
- Neural Networks
    - Mimics neurons
        - Takes inputs
        - Multiply by weights
        - Apply activation functions
        - Stack multiple layers.
    - Reallly good for nonlinear patterns

And even more importantly:
- Studying class imbalance
    - Class imbalance is when there is one side that is very imbalanced, this imbalance can allow for the model to get away with predicting the majority or the bigger probability, making it good at finding a non -defaulter rather than being good at finding a defaulter.
- Understanding feature importance
- Learning ROC/AUC
    - ROC Curve  (ROC = Receiver Operating Characteristic)
        - It plots: True Positive Rate (Recall) vs False Positive Rate
    - AUC (Area Under the Curve)
        - Tells you how well the model separates the two classes.
- Learning confusion matrices

|            | Predicted No | Predicted Yes |
| ---------- | ------------ | ------------- |
| Actual No  | TN           | FP            |
| Actual Yes | FN           | TP            |

- Learning precision vs recall tradeoffs
    - Precission: Of the people predicted to default, how many actually defaulted
    - Recall: Of all actual defaulters: How many did we catch?
        - In credit risk, recall is usually very important.

````python
target_col = next(
    col for col in ["default payment next month", "default_payment_next_month", "Y"]
    if col in df.columns
)

total_accounts = len(df)
defaulted_accounts = (df[target_col] == 1).sum()
default_rate = (defaulted_accounts / total_accounts) * 100

print(f"Total accounts: {total_accounts}")
print(f"Defaulted accounts: {defaulted_accounts}")
print(f"Default rate: {default_rate:.2f}%")
````

Paying attention to the fact that the default rate is only around 20% it presents the possibility of having a bad model that trys predicting who won't default and because the presentage of non defaults is 78% it will think that it is highly accurate but in reality the model is just useless.

When one class is smaller than the other the model may become biased toward the majority class

It learns:

    “I can be mostly right just by predicting no default.”

### Why This Is Dangerous in Credit Risk

Missing a defaulter (False Negative) means:

- You approve a risky customer
- They default
- You lose money

So for banks:
- Recall for defaulters is often more important than raw accuracy.

The problem isn’t just:

    “We don’t have enough examples.”

The problem is:

    The model is incentivized to ignore the minority class unless we intervene.

### How We Fix It

Use better metrics:
- Recall
- Precision
- F1 Score
- ROC/AUC

Adjust class weights:
``` python
    LogisticRegression(class_weight='balanced')
```
