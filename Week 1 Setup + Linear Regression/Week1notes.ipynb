## Topic 1: Training Linear Regression

### What is Linear Regression?:

- Lineear regression is a statistical technique for modeling the relationship between an output variable and one or more input variables
    - Fitting a line between data points to make predictions between unkown data.
    - In y = mx + b; m = B1 (beta 1), b = B0 (Beta 0) and in machine learning beta 0 is the bias.

### Residual:
- A residual is the differences between the points and the line, the difference between the data point and the prediction the line will produce.
